```python
import networkx as nx
import numpy as np
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

```python
# formula of node_color= "cyan" : kp = 2q  , p = nodes
k = 6
nodes = 500
G =nx.random_regular_graph(k, nodes)
nx.draw(G,node_size=100,font_size=8, node_color= "cyan",  with_labels=True)
```

```python
discoonectivity = 0
for i in range(100000):
  u = random.randint(0, G.number_of_nodes())
  v = random.randint(0, G.number_of_nodes())
  if u != v and G.has_edge(u,v):
    G.remove_edge(u, v)
  if G.number_of_edges() <= 1251:
    discoonectivity += 1

print("discoonectivity of the graph: " + str(discoonectivity/100000))
print("number of edges: " + str(G.number_of_edges()))
```

```python
d1 = 4
d2 = 6
n = 500

n1 = int((n * d2) / (d1 + d2))
n2 = int((n * d1) / (d1 + d2))

nodes_a = []
nodes_b = []
for i in range(1, n1 + 1):
  nodes_a += [i for x in range(d1)]

for i in range(n1 + 1, n1 + n2 + 1):
  nodes_b += [i for x in range(d2)]
```

```python
# Draw RSRBG graph
nodes_check = []
check_edge = {}
A = nx.Graph()
while (len(nodes_a) != 0):

  choose_a_index = random.randint(0, len(nodes_a))
  choose_b_index = random.randint(0, len(nodes_b))

  remove_a = nodes_a[ choose_a_index - 1 ]
  remove_b = nodes_b[ choose_b_index - 1 ]

  if nodes_check.count(remove_a) != True:
    nodes_check.append(remove_a)
    G.add_node(remove_a)
  if nodes_check.count(remove_b) != True:
    nodes_check.append(remove_b)
    G.add_node(remove_b)

  if not A.has_edge(remove_a, remove_b):
    A.add_edge(remove_a, remove_b)
    nodes_a.pop( choose_a_index - 1 )
    nodes_b.pop( choose_b_index - 1 )


color_map = []

for node in G:
  if node < n1 +1:
        color_map.append('yellow')
  else:
        color_map.append('pink')

nx.draw(A,node_size=100,font_size=8, node_color= "cyan",  with_labels=True)
```

```python
print("node connectivity: " + str(nx.node_connectivity(A, s=None, t=None, flow_func=None)))
print("edge connectivity: " + str(nx.edge_connectivity(A, s=None, t=None, flow_func=None, cutoff=None)))
print('Algebraic Connectivity: ' + str(nx.algebraic_connectivity(A)))
```
